```python
import os
import json
import random

# Đường dẫn dữ liệu
INPUT_FILE = "../data/jsonl/vihallu-train.jsonl"
TRAIN_OUT = "../data/processed/vihallu-train-split.jsonl"
VAL_OUT = "../data/processed/vihallu-val-split.jsonl"

# Tỷ lệ chia và seed
SPLIT_RATIO = 0.1
SEED = 42

# Tạo thư mục output nếu chưa có
os.makedirs("../data/processed", exist_ok=True)
print("Thư mục processed đã sẵn sàng.")
```

```python
def split_dataset(input_path, train_path, val_path, split_ratio=0.1, seed=42):
    with open(input_path, "r", encoding="utf-8") as f:
        data = [json.loads(line) for line in f]

    # Trộn ngẫu nhiên
    random.seed(seed)
    random.shuffle(data)

    # Chia train / val
    n_total = len(data)
    n_val = int(n_total * split_ratio)
    val_data = data[:n_val]
    train_data = data[n_val:]

    # Ghi ra file JSONL
    with open(train_path, "w", encoding="utf-8") as f_train:
        for item in train_data:
            f_train.write(json.dumps(item, ensure_ascii=False) + "\n")

    with open(val_path, "w", encoding="utf-8") as f_val:
        for item in val_data:
            f_val.write(json.dumps(item, ensure_ascii=False) + "\n")

    print(f"Tổng số mẫu: {n_total}")
    print(f"Train: {len(train_data)} mẫu → {train_path}")
    print(f"Val: {len(val_data)} mẫu → {val_path}")
```

```python
split_dataset(INPUT_FILE, TRAIN_OUT, VAL_OUT, SPLIT_RATIO, SEED)
```
